# 活性化関数
- tanh関数
- シグモイド関数

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-5, 5, 0.1)

def relu(x):
    return np.maximum(0, x)

tanh = np.tanh(x)
sigm = 1 / (1 + np.exp(-x))
re   = relu(x)
```

```python
plt.title("Activation function")
plt.xlabel("x")
plt.ylabel("y")
plt.plot(x, tanh, label="tanh")
plt.plot(x, sigm, label="sigmoid")
plt.plot(x, re,   label="reLU")
plt.legend()
plt.show()
```
